#### Groupping sequences


```python
def get_distance_matrix(str_list):
    """ Construct a levenshtein distance matrix for a list of strings"""
    dist_matrix = np.zeros(shape=(len(str_list), len(str_list)))
    #t0 = time()
    #print "Starting to build distance matrix. This will iterate from 0 till ", len(str_list) 
    for i in range(0, len(str_list)):
        #print i
        for j in range(i+1, len(str_list)):
                dist_matrix[i][j] = lev.distance(str_list[i], str_list[j]) 
    for i in range(0, len(str_list)):
        for j in range(0, len(str_list)):
            if i == j:
                dist_matrix[i][j] = 0 
            elif i > j:
                dist_matrix[i][j] = dist_matrix[j][i]
    #t1 = time()
    #print "took", (t1-t0), "seconds"
    return dist_matrix
```

```python
vc = train.x.value_counts()
grp = 0

# MUTATION THRESHOLD
M_THRESHOLD = 10
# INSERTION DELETION THRESHOLD
D_THRESHOLD = 3

for k in range(len(vc)):
    c = vc.index[k]
    # SUBSET OF TRAIN DATA WITH SAME PROTEIN LENGTH PLUS MINUS D_THRESHOLD
    tmp = train.loc[(train.x>=c-D_THRESHOLD)&(train.x<=c+D_THRESHOLD)&(train.group==-1)]
    if len(tmp)<=1: break
    # COMPUTE LEVENSTEIN DISTANCE
    x = get_distance_matrix(list(tmp.protein_sequence))
    # COUNT HOW MANY MUTATIONS WE SEE
    mutation = []
    for kk in range(1,M_THRESHOLD+1):
        mutation.append( len( np.unique( np.where( x==kk )[0] ) ) )
    # FIND RELATED ROWS IN TRAIN WITH M_THRESHOLD MUTATIONS OR LESS
    y = np.unique( np.where( (x>0)&(x<=M_THRESHOLD) )[0] )
    seen = []
    for j in y:
        if j in seen: continue
        i = np.where( np.array(x[j,])<=M_THRESHOLD )[0]
        seen += list(i)
        idx = tmp.iloc[i].index
        train.loc[idx,'group'] = grp
        grp += 1
    ct = vc.iloc[k]
    ct2 = len(tmp)
    print(f'k={k} len={c} ct={ct} ct2={ct2} dist_ct={mutation}')
    #if k==9: break
```

```python
def groups_dict(df):
    """split df into their groups
    """
    #unqiue names
    uniquenames = df.group.unique()
    DataFrameDict = {}
    n = 0

    for elem in uniquenames:
        DataFrameDict[str(elem)] = df[df['group'] == elem]
        n += 1
    
    for key, data in DataFrameDict.items():
        #get index for base seq
        idx = data[data['tm'] == data['tm'].min()].index[0]
        base = data.protein_sequence[idx]
        data['similar_ratio'] = data['protein_sequence'].apply(lambda x: lev.ratio(base, x))
    
    
    
    return DataFrameDict
```

```python
def mutation_helper(base, query_seq):
    """
    helper function
    return: list of mutations
    """
    muts_list = []
    for mut,first,sec in lev.editops(base, query_seq):
        #print(mut,first,sec)
        if mut == 'delete':
            muts_list.append(str(mut)[0:3]+base[first])
        elif mut == 'replace':
            muts_list.append(query_seq[sec]+str(mut)[0:3]+base[first])
        else:
            if first > len(base):
                muts_list.append(query_seq[sec]+str(mut)[0:3]+base[len(base)-1])
            elif first == 0:
                muts_list.append(query_seq[sec]+str(mut)[0:3]+base[first])
            else:
                muts_list.append(query_seq[sec]+str(mut)[0:3]+base[first-1])
    muts_list = set(muts_list)
    return list(muts_list)
def mutations_groups(df_dict):
    """
    add mutations type columns to dataframe groups
    """
    dictionary = df_dict.copy()
    for key, data in dictionary.items():
        #get index for base seq
        idx = data[data['tm'] == data['tm'].min()].index[0]
        base = data.protein_sequence[idx]
        data['mutations'] = data['protein_sequence'].apply(lambda x: mutation_helper(base, x))
        data['total_mutations'] = data['mutations'].apply(lambda x: len(x))
        
    return dictionary
```

#### Use amino acid properties into meaningful features
- calculate the avg Molecular weight and avg residue weight
- avg Solvent accessible surface area (SASA) of proteins, has always been considered as a decisive factor in protein folding and stability studies. It is defined as the surface characterized around a protein by a hypothetical centre of a solvent sphere with the van der Waals contact surface of the molecule.

```python
mutations_df.head(2)
```

```python
#create mapping to use Amino acid
dict_map = {}
for idx in aa_props.index:
    dict_map[idx] = (aa_props['Molecular Weight'][idx], aa_props['Residue Weight'][idx], aa_props['SASA'][idx])
```
